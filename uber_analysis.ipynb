```python
import pandas as pd

df = pd.read_csv("uber-raw-data-apr14.csv")

df.head()
```

```python
df['Date/Time'] = pd.to_datetime(df['Date/Time'])

df['hour'] = df['Date/Time'].dt.hour
df['day'] = df['Date/Time'].dt.day
df['weekday'] = df['Date/Time'].dt.weekday
```

```python
import matplotlib.pyplot as plt

df['hour'].value_counts().sort_index().plot(kind='bar')

plt.title("Uber Rides by Hour")

plt.xlabel("Hour of Day")

plt.ylabel("Number of Rides")

plt.show()
```

```python
df['weekday'].value_counts().sort_index().plot(kind='bar')

plt.title("Uber Rides by Weekday")

plt.show()
```

## Key Insights

- Ride demand peaks during evening hours.
- Weekends show increased ride activity.
- Late-night ride demand remains consistent in urban areas.
- Ride distribution varies significantly across different hours of the day.
